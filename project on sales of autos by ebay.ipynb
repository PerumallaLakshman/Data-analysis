```python
# 4) What is the Average price of a vehicle by brand as well as vehicle type.Use heatmap to explain this
```

```python
datax=datac.groupby(["vehicleType","brand"])
datax["price"].mean()
```

```python
avg_price=pd.DataFrame(pd.crosstab(index=datac["brand"],columns=datac["vehicleType"],values=datac["price"],aggfunc=["mean"]))
avg_price.fillna(0,inplace=True)
avg_price
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(data=avg_price,annot=True,fmt=".4g")
plt.title("Average price of a vehicle by brand as well as vehicle type")
plt.show()
```
